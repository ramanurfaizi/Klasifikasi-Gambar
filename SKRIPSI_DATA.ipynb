```python
# mengimpor tiga pustaka di Python yaitu NumPy, Matplotlib, dan Pandas
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Membaca dataset dari file CSV dengan menggunakan Pandas.
dataset = pd.read_csv("837-pekerjaan.csv", sep=';')
dataset
```

```python
# melakukan encoding (penyandian) pada data kategori atau label menjadi bilangan bulat.
from sklearn.preprocessing import LabelEncoder
```

```python
# membuat objek LabelEncoder dari modul scikit-learn
enc = LabelEncoder()
```

```python
# melakukan encoding pada kolom 'Jenis Pekerjaan' dan 'Kelas' dalam dataset menggunakan suatu objek encoder, yang kemudian dapat
# digunakan untuk mengubah nilai-nilai kategorikal menjadi nilai numerik.
dataset['Jenis Pekerjaan'] = enc.fit_transform(dataset['Jenis Pekerjaan'].values)
dataset['Kelas'] = enc.fit_transform(dataset['Kelas'].values)
```

```python
dataset
```

```python
# Menghapus kolom 'Jenis Pekerjaan' dari dataset dan menyimpannya ke dalam variabel
atr_dataset = dataset.drop(columns = 'Kelas')
atr_dataset
```

```python
# Mengambil kolom "Kelas" dari DataFrame dataset dan menyimpannya dalam variabel cls_dataset.
cls_dataset = dataset ['Kelas']
cls_dataset
```

```python
#  digunakan untuk mengimpor modul-modul yang dibutuhkan dari scikit-learn (sebuah pustaka machine learning di Python)
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier
```

```python
# membagi dataset menjadi data pelatihan (train) dan data uji (test) menggunakan train_test_split, dan
# kemudian melatih model Decision Tree pada data pelatihan (xtrain, ytrain)
xtrain, xtest, ytrain, ytest = train_test_split(atr_dataset, cls_dataset, test_size=0.4, random_state=1)
tree_dataset = DecisionTreeClassifier(random_state=1)
tree_dataset.fit(xtrain, ytrain)
```

```python
#  melakukan prediksi menggunakan model pohon keputusan (C4.5) pada data pengujian (xtest), menghitung dan menampilkan confusion matrix,
# classification report (termasuk akurasi), serta tingkat akurasi dalam persentase.

y_pred = tree_dataset.predict(xtest)
cm = confusion_matrix(ytest, y_pred)
print("Confusion Matrix")
print(cm)
akurasi = classification_report(ytest, y_pred, zero_division=1)
print("Tingkat Akurasi Algoritma C4.5")
print("Akurasi :", akurasi)
akurasi = accuracy_score(ytest, y_pred)
print("Tingkat Akurasi : %d persen" %(akurasi*87))
```

```python
# menghasilkan file DOT yang merepresentasikan struktur pohon keputusan (decision tree) dari model yang telah dilatih menggunakan dataset (tree_dataset).
from sklearn.tree import export_graphviz
export_graphviz(tree_dataset, out_file="Tree_Kelas.dot", class_names=["Pegawai Buruh", "Pegawai Negeri", "Pegawai Swasta", "Polisi","Pertanian","Pedagang","Supir Bis","Trans"],
                feature_names=atr_dataset.columns, impurity=False, filled=True)
```

```python
# membaca dan menampilkan visualisasi pohon keputusan (dalam format DOT) yang disimpan dalam file "Tree_Kelas.dot" menggunakan perpustakaan Graphviz.

import graphviz

with open("Tree_Kelas.dot") as fig:
  dot_graph = fig.read()
  graph = graphviz.Source(dot_graph)

graph.view()
```
